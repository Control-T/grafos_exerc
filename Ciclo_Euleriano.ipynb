##Um tour de Euler é um ciclo onde se parte de um vértice inicial e segue por todas as arestas do grafo, sem repetí-las, até voltar ao vértice de origem.
##Pelo fato das figura dada possuir poucos vértices é fácil constatar que todos possuem grau par.
##Pelo teorema 8.2 (Euler-Hierholzer) um grafo G=(V,E) não orientado e conexo possui um ciclo euleriano se e somente se todos os seus vértices tiverem grau par.

#Algoritmo proposto para a solução do problema.

```python
import networkx as nx
import matplotlib.pyplot as plt

G=nx.Graph()

G.add_nodes_from(['1','2','3','4','5','6','7','8'])

G.add_edges_from([('1','2'),('1','3'),('1','6'),('1','8'),
                  ('2','4'),('2','5'),('2','7'),
                  ('3','8'),
                  ('4','7'),
                  ('5','7'),
                  ('6','8'),
                  ('7','8')])

tour_euler = nx.eulerian_circuit(G, '1')
arestas = list(tour_euler)

nx.draw(G, pos=nx.circular_layout(G), with_labels=True, font_weight='bold', font_color="w", node_color='b')
plt.draw()
print('Tour de Euler a seguir:', arestas, '\n\n')
```
